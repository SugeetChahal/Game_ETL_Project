```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "Resources/vgsalesGlobale.csv"
game_df = pd.read_csv(csv_file)
game_df.head()
```

```python
new_game_df = game_df[['Name', 'Publisher', 'Year', 'Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].copy()
new_game_df.head()
```

```python
csv_file = "Resources/Video_Games.csv"
game_2_df = pd.read_csv(csv_file)
game_2_df.head()
```

```python
new_game_2_df = game_2_df[['Name', 'Publisher', 'Year_of_Release', 'Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].copy()
new_game_2_df.head()
```

```python
new_game_2_df.loc[~np.isfinite(new_game_2_df['Year_of_Release'])]
```

```python
new_game_2_df.Year_of_Release = new_game_2_df.Year_of_Release.astype(int)
```

```python
csv_file = "Resources/steam.csv"
game_3_df = pd.read_csv(csv_file)
game_3_df.head()
```

```python
new_game_3_df = game_3_df[['name', 'publisher', 'release_date', 'genres', 'average_playtime', 'median_playtime', 'owners', 'price']].copy()
new_game_3_df.head()
```

```python
new_game_3_df[['Min_Owners','Max_Owners']] = new_game_3_df.owners.str.split("-",expand=True,)
```

```python
new_game_3_df.Min_Owners = new_game_3_df.Min_Owners.astype(int)
new_game_3_df.Max_Owners = new_game_3_df.Max_Owners.astype(int)
```

```python
new_game_3_df.head()
```

```python
new_game_3_df = new_game_3_df[['name', 'publisher', 'release_date', 'genres', 'average_playtime', 'median_playtime', 'Min_Owners', 'Max_Owners', 'price']].copy()
new_game_3_df.head()
```

```python
new_game_3_df.release_date = new_game_3_df.release_date.astype(np.datetime64)
```

```python
new_game_3_df['release_date'].astype(np.datetime64)
```

```python
rds_connection_string = "postgres:1379@localhost:5432/game_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_game_df.to_sql(name='game1', con=engine, if_exists='append', index=False)
```
